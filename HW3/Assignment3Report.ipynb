2. Both conversations are saved in the same .db file. Explain how the session_id prevented Context Pollution; i.e., mixing the two topics.

Session_id was a variable that stored each user’s session history with different ids so that it makes sure that each user’s chat history is local to only those sessions. For example, in this experiment the two types of sessions users could have were either  science facts or American history. However, since each of those 2 sessions had a unique id, the user’s questions (context memory) cannot be remembered from external sessions. I saw this in the experiment when I went to the American History session; it had no idea that I said anything about a sun in the Science Facts one. This prevents Context Pollution, the act of different sessions remembering the other’s information, and in the real-world, can prevent breach of privacy/confidentiality if a session holds such discreet information.
